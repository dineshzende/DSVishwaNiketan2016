```python
import pandas as pd
f = open("C:/MoviesData/ratings.dat")
user_id=[]
movie_id=[]
rating=[]
data = f.readlines()
for i in data:
    data_list = i.split('::')
    #print data_list
    user_id.append(data_list[0])
    movie_id.append(data_list[1])
    rating.append(data_list[2])

df = pd.DataFrame({'user_id': user_id,
                 'movie_id': movie_id,'ratings':rating})
df.sort_values(by=["ratings"],ascending=[False],inplace=True)

df
```

```python
# Finding out Highly rated movies - Genere
import pandas as pd
import numpy as np
rating_df = pd.read_csv("C:/MoviesData/ratings.dat",sep='::',engine='python',names=['user_id','movie_id','ratings','timestamp'])
movie_df = pd.read_csv("C:/MoviesData/movies.dat",sep='::',engine='python',names=['movie_id1','movie_name','gener'])
inner_joined = pd.merge(rating_df, movie_df, how='inner', left_on='movie_id', right_on='movie_id1')
#inner_joined.sort_values(by=['ratings'],ascending=[False],inplace=True)
inner_joined.drop('timestamp',axis='columns',inplace=True)
inner_joined.drop('movie_id1',axis='columns',inplace=True)

grouped = inner_joined.pivot_table(values='ratings',index='gener',aggfunc=np.mean)
print(type(grouped))
grouped_df = grouped.to_frame(name=None)
grouped_df.sort_values(by=['ratings'],ascending=[False],inplace=True)
grouped_df[:5]
```

```python
# Plotting distribution of age of users

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Read users.dat file
users_df = pd.read_csv("C:/MoviesData/users.dat",sep='::',engine='python',names=['user_id','gender','age','occupation','zip-code'])
#Group by user-group and append column of count of entries at the end
grouped = users_df.groupby(['age'])['age'].count()
#Convert Series to DataFrame 
grouped_df = grouped.to_frame(name=None)
#grouped_df
#Convert index as a column and rename 'age' to 'age_count'
grouped_df.rename(index={'age': 'age_group'},columns={'age': 'age_count'},inplace=True)
#Reset index 
grouped_df.reset_index(level=0, inplace=True)
print(grouped_df)
print(type(grouped_df))
#Plot Histogram of the count
grouped_df['age_count'].hist(bins=10)
#plt.hist(grouped_df['age'],bins=10)
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend()
#plt.hist(grouped_df['user_age'],bins=20)
plt.show()
```
